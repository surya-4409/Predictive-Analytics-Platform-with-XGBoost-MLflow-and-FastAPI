```python
import pandas as pd
from shopsense.data_generator import generate_customers, generate_transactions, generate_events
from shopsense.features import compute_rfm_features, compute_behavioral_features, compute_transaction_features, build_master_feature_table
from shopsense.eda import test_premium_vs_nonpremium_aov, test_channel_churn_association, test_revenue_seasonality

customers_df = generate_customers(5000, random_state=42)
transactions_df = generate_transactions(customers_df, random_state=42)
events_df = generate_events(customers_df, random_state=42)
snapshot_date = '2023-12-31'

# Feature Engineering
rfm = compute_rfm_features(transactions_df, snapshot_date)
behaviors = compute_behavioral_features(events_df, transactions_df, snapshot_date)
txns = compute_transaction_features(transactions_df, snapshot_date)
master_df = build_master_feature_table(customers_df, rfm, behaviors, txns)
display(master_df.head())

# Statistical Testing
print("\n--- Hypothesis Test: AOV ---")
display(test_premium_vs_nonpremium_aov(transactions_df, customers_df))

print("\n--- Hypothesis Test: Seasonality ---")
display(test_revenue_seasonality(transactions_df))
```
